Zamieniamy również kolumnę "Type" na wartości binarne

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
dicts = {}

label.fit(sales_feature.Type.drop_duplicates()) 
dicts['Type'] = list(label.classes_)
sales_feature.Type = label.transform(sales_feature.Type)
```

```python
sales_feature.head(10)
```

```python
# Do zrobienia:
# Znalezc numery tygodnia, dnia tygodnia, miesiaca, kwartalu
# znalezc srednie tygodniowe, miesieczne, kwartlane
```

```python
month = sales_feature.Date.apply(lambda d: d.month)
quarter = sales_feature.Date.apply(lambda d: d.quarter)
day_of_week = sales_feature.Date.apply(lambda d: d.weekday())
```

```python
sales_feature['month'] = month
sales_feature['quarter'] = quarter
sales_feature['day_of_week'] = day_of_week
```

```python
monthly_sales = sales_feature[['month', 'Weekly_Sales']].groupby(['month']).mean()
monthly_sales.columns = ['monthlySales']
q_sales = sales_feature[['quarter', 'Weekly_Sales']].groupby(['quarter']).mean()
q_sales.columns = ['q_sales']
dow_sales = sales_feature[['day_of_week', 'Weekly_Sales']].groupby(['day_of_week']).mean()
dow_sales.columns = ['dow_sales']
```

```python
sales_feature = sales_feature.merge(monthly_sales.reset_index(), left_on="month", right_on="month")\
    .merge(q_sales.reset_index(), left_on="quarter", right_on="quarter")\
    .merge(dow_sales.reset_index(), left_on='day_of_week', right_on='day_of_week')
```

```python
sales_feature.fillna(0, inplace=True)
```
